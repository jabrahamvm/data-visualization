```python
n_rows = 0
```

```python
for row in january:
    times = row[-1]
    d_format = row[0].strftime('%Y-%m-%d')
    insert_query = f"INSERT INTO registro_denuncias (fecha,idDelito,idEstado,idPoblacion) VALUES ('{d_format}',{row[1]},{row[2]},{row[3]});"
    for i in range(times): 
        mysql.execute_query(connection,insert_query)
        n_rows += 1
```
